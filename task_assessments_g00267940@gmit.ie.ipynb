### Step 3: Exception statement added inside another

It was possible to modify examples found through web searches for Step 1 and Step 2 listed above. There is no link for Step 3 as the idea came from using if statements inside one another. The author applied the same logic to exception statements without assistance from web searches.

#### Note: 
Including an exception statement inside another exception statement meant it was then possible to differentiate between a string, integer and float data type. The code in its current structure cannot differentiate between a list inside a list and a string.

```python
# Create empty list
task_list = []

# Input list size
n_1 = int(input("Enter the list size: "))

# populate list using for list for 'x' inputs 'n' times
for x in range(n_1):
    x = input("")
    
    #exception handling statement so code can identify different data types
    try:  
        task_list.append(int(x))
    except ValueError:
        
        #statement inside a statement so code can differentiate between float and string elements
        try:
            task_list.append(float(x))
        except ValueError:
            task_list.append(x)
print(task_list)
```

### Step 4: Create a function - Counts() 
A function that takes each element in tasks_list() and returns a dictionary of each list input as a key and the number of times it appears as the value {key: value}.

[3] https://www.geeksforgeeks.org/counting-the-frequencies-in-a-list-using-dictionary-in-python/ - online python tutorial for counting frequencies in lists

```python
# Define function called Counts that takes elements in task_list as inputs
def counts_1(task_list): 
      
   # Creating an empty dictionary  
   count_1 = {} 
   for i in (task_list):
    # The get () method returns the value for the specified key if key is in dictionary
    count_1[i] = count_1.get(i, 0) + 1
   return count_1 

# Driver function 
if __name__ == "__main__":  
    print(counts_1(task_list))
```

# Fundamentals of Data Analysis: Task Assessment 2

```python
#import the required libraries
import numpy as np
import pandas as pd
import collections
import matplotlib.pyplot as plt
```

```python
#Enter the number of dice and the amount of times they are to be rolled
k = int(input("Enter the number of dice: "))
n_2 = int(input("Enter many times the dice are to be rolled: "))
```

```python
#dicerolls function using k, n parameters
def dicerolls(k, n_2):
    
    #use random module to generate array with elements in range of 1 to 6
    diceroll_rng = np.random.default_rng().integers(1,7, size=(k, n_2))
    #sum the rows (k value) if required so that it is a 1-d array
    diceroll_rng = np.sum(diceroll_rng, axis=0)
  
    return diceroll_rng

diceroll = dicerolls(k, n_2)
```

```python
#variable diceroll is a 1-d array that returns dice roll sum values
diceroll
```

```python
# counts function that takes elements in diceroll as inputs and 
# returns a dictionary with each element and its frequency
def counts_2(diceroll):
    
    count_2 = {} 
    for i in diceroll:
        
        count_2[i] = count_2.get(i, 0) + 1

    return count_2

# Driver function 
if __name__ == "__main__":  
    print(counts_2(diceroll))
```

```python
# Generate a histrogram showing the frequency of rolled values for k number of dice
plt.hist(diceroll, bins=(6+(5*(k-1))), color='grey')
plt.title("Dicerolls: frequency distribution")
plt.xlabel("Value of the dice roled")
plt.ylabel("Number of times value is recorded")
plt.show()
```

```python
# Convert the array to a pandas DataFrame
df_diceroll = pd.DataFrame(diceroll)
```

```python
# Use the describe function to perform statistical analysis on the data
df_diceroll.describe()
```

## Conclusion:

It is clear from the histogram that the possiblility of rolling the lowest or highest value on the x axis is significantly lower than rolling a a value in the middle of the distribution. This is to be expected as rolling the lowest or the highest value woudld require all dice recording either the lowest combined or highest combined value which is going to be least likely. Statistical analysis shows that the mean and the 50th percentile are very close which is what would be expected for a normally distributed dataset.

```python
#Parameters provided in task assessment
coin_toss = 100
number_of_tests = 1000

#Heads value is 0 at the beginning
heads = 0

#List to be populated with coin_toss values
heads_count = []

#Define function to execute 'test_size' times
def test_thou(number_of_tests):
    
    #Clear the list to begin
    list.clear(heads_count)
    
    x= 0
    
    for x in range(number_of_tests):
        
        #Nested function so that results can be grouped per 100 coin toss
        #https://stackabuse.com/python-nested-functions/ guidance from website for nesting functions
        def test_hund(coin_toss):

            heads = 0

            for i in range(coin_toss):

                heads_lands = random.randint(1,2)
                if heads_lands == 1:
                    heads += 1

            heads_count.append(heads)
        test_hund(coin_toss)
        
        x += 1
        
test_thou(number_of_tests)
```

```python
heads_count
```

Step 3) Create a dictionary that returns the numbers of times in the number of tests in which heads is recorded.

```python
heads_dict = {}

for i in heads_count:

    heads_dict[i] = heads_dict.get(i, 0) + 1
    heads_dict = dict(sorted(heads_dict.items()))
    
print(heads_dict)
```

Step 4) Determine the number of bins required to accurately display the data.

```python
# Create a list with key values from count dictionary
keys = list(heads_dict.keys())

# Subtract the max from the min key value and add 1 as this value is part of the dataset
bin_value = (max(keys) - min(keys))+1
```

Step 5) Generate a histrogram showing the numbers of times in the number of tests in which heads was recorded.

```python
plt.hist(heads_count, bins=bin_value)
plt.xlabel("Number of times heads lands out of 100")
plt.ylabel("Number of tests")
plt.show()
```

Step 6) Normal Distributions.

It is clear from the histogram that the data is normally distributed. By calculating the mean and the standard deviation of the dataset it is possible to plot a curve to fit the data. The code for fitting a normal distribution to a dataset was found at https://stackoverflow.com/questions/20011122/fitting-a-normal-distribution-to-1d-data with minor changes to match this dataset.

```python
# Fit a normal distribution to the data:
mean, std_dev = norm.fit(heads_count)
```

```python
# Plot the histogram.
plt.hist(heads_count, bins=bin_value, density=True, alpha=0.6, color='black')

# Plot the Distribution.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, bin_value)
p = norm.pdf(x, mean, std_dev)

plt.plot(x, p, 'red', linewidth=2)
title = "Fit results: mean = %.2f,  std_dev = %.2f" % (mean, std_dev)
plt.title(title)

plt.show()
```
